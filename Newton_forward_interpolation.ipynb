```python
import numpy as np

print("Enter number of data points: ")
n = int(input())

x = np.zeros(n, dtype=float)
y = np.zeros(n, dtype=float)

print("Enter data points (x y, one per line): ")
for i in range(n):
    while True:
        try:
            data_input = list(map(float, input(f"Data point {i+1} (x y): ").split()))
            if len(data_input) != 2:
                print("Please enter exactly two numbers (x y).")
                continue
            x[i] = data_input[0]
            y[i] = data_input[1]
            break
        except ValueError:
            print("Invalid input. Please enter space-separated numbers.")

# Initialize the forward difference table
d = np.zeros((n, n), dtype=float)
for i in range(n):
    d[i, 0] = y[i]

# Calculate forward differences
for j in range(1, n):
    for i in range(n - j):
        d[i, j] = d[i + 1, j - 1] - d[i, j - 1]

print("Enter value of x for interpolation: ")
xp = float(input())

h = x[1] - x[0] # Assuming equally spaced x values
p = (xp - x[0]) / h

sum_val = y[0]
term = 1.0

# Newton's Forward Interpolation Formula
for i in range(1, n):
    term = term * (p - (i - 1)) / i
    sum_val += term * d[0, i]

print(f"\nInterpolated value at x = {xp}: {sum_val:.6f}")
```
