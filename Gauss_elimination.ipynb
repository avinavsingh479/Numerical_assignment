```python
import numpy as np

print("Enter number of variables: ")
n = int(input())

# Initialize augmented matrix A (n x n+1)
A = np.zeros((n, n + 1), dtype=float)

print("Enter augmented matrix (e.g., '1 2 4' for first row with 2 variables, where 4 is the constant term):")
for i in range(n):
    while True:
        try:
            # Read n+1 values for each row: n coefficients and 1 constant term
            row_input = list(map(float, input(f"Row {i+1}: ").split()))
            if len(row_input) != n + 1:
                print(f"Please enter exactly {n+1} numbers for row {i+1}.")
                continue
            for j in range(n + 1):
                A[i, j] = row_input[j]
            break
        except ValueError:
            print("Invalid input. Please enter space-separated numbers.")

# Gaussian Elimination (Forward Elimination)
for i in range(n):
    # Find pivot element (largest absolute value in the current column below the diagonal)
    pivot_row = i
    for k in range(i + 1, n):
        if abs(A[k, i]) > abs(A[pivot_row, i]):
            pivot_row = k

    # Swap current row with pivot row
    A[[i, pivot_row]] = A[[pivot_row, i]]

    # Check for division by zero (singular matrix)
    if A[i, i] == 0.0:
        print("Error: Division by zero encountered. The matrix may be singular or poorly conditioned.")
        exit()

    for j in range(i + 1, n):
        ratio = A[j, i] / A[i, i]
        for k in range(n + 1):
            A[j, k] -= ratio * A[i, k]

# Back Substitution
x = np.zeros(n, dtype=float)
for i in range(n - 1, -1, -1):
    x[i] = A[i, n]
    for j in range(i + 1, n):
        x[i] -= A[i, j] * x[j]
    if A[i, i] == 0:
        print("Error: Division by zero during back substitution. This should not happen if forward elimination was successful.")
        exit()
    x[i] /= A[i, i]

print("\nSolution:")
for i in range(n):
    print(f"{x[i]:.6f} ", end="")
print()
```
