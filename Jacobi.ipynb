```python
import numpy as np

def jacobi_method():
    n = int(input("Enter number of variables: "))

    print("Enter augmented matrix (row by row, space-separated values):")
    a = np.zeros((n, n + 1), dtype=float)
    for i in range(n):
        while True:
            try:
                row_input = list(map(float, input(f"Row {i+1}: ").split()))
                if len(row_input) != n + 1:
                    print(f"Please enter exactly {n + 1} values for this row.")
                    continue
                a[i] = row_input
                break
            except ValueError:
                print("Invalid input. Please enter space-separated numbers.")

    iter = int(input("Enter iterations: "))

    x = np.zeros(n, dtype=float)  # Initial guess x0
    x1 = np.zeros(n, dtype=float)

    for k in range(iter):
        for i in range(n):
            sum_val = a[i, n] # The constant term b_i
            for j in range(n):
                if i != j:
                    sum_val -= a[i, j] * x[j]
            x1[i] = sum_val / a[i, i]

        # Update x with the new approximations x1
        x = np.copy(x1)

    print("\nSolution (x values): ")
    for i in range(n):
        print(f"x[{i}] = {x[i]:.6f}")

jacobi_method()
```
