```python
import numpy as np

print("Enter number of variables: ")
n = int(input())

# Initialize A (n x n) and b (n x 1)
A = np.zeros((n, n), dtype=float)
b = np.zeros(n, dtype=float)

print("Enter augmented matrix (e.g., '4 1 9' for first row with 2 variables, where 9 is the constant term):")
for i in range(n):
    while True:
        try:
            # Read n+1 values for each row: n coefficients and 1 constant term
            row_input = list(map(float, input(f"Row {i+1}: ").split()))
            if len(row_input) != n + 1:
                print(f"Please enter exactly {n+1} numbers for row {i+1} (coefficients + constant term).")
                continue
            for j in range(n):
                A[i, j] = row_input[j]
            b[i] = row_input[n] # The last element is the constant term
            break
        except ValueError:
            print("Invalid input. Please enter space-separated numbers.")

print("Enter iterations: ")
iterations = int(input())

# Initial guess for x, all zeros
x = np.zeros(n, dtype=float)

# Gauss-Seidel iteration
for k in range(iterations):
    for i in range(n):
        sum_val = b[i]
        for j in range(n):
            if i != j:
                sum_val -= A[i, j] * x[j] # Use the most recently updated x[j]
        if A[i, i] == 0:
            print(f"Error: Diagonal element A[{i},{i}] is zero. Cannot perform Gauss-Seidel iteration.")
            exit() # or handle error appropriately
        x[i] = sum_val / A[i, i] # Update x[i] immediately

print("\nSolution after", iterations, "iterations:")
for i in range(n):
    print(f"{x[i]:.6f} ", end="")
print()
```
