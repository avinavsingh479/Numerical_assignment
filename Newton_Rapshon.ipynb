```python
def f(x):
    return x*x*x - x - 2

def df(x):
    return 3*x*x - 1

def newton_raphson():
    initial_guess = float(input("Enter initial guess: "))
    iterations = int(input("Enter iterations: "))

    x = initial_guess

    for i in range(iterations):
        # Check for division by zero (derivative being zero)
        derivative_val = df(x)
        if derivative_val == 0:
            print("Error: Derivative is zero. Cannot proceed with Newton-Raphson.")
            return
        x = x - f(x) / derivative_val

    print(f"Root = {x:.6f}")

newton_raphson()
```
