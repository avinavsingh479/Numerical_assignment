```python
def f(x):
    return x*x*x - x - 2

def bisection_method():
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    n = int(input("Enter iterations: "))

    c = 0.0 # Initialize c

    for i in range(n):
        c = (a + b) / 2
        if f(c) == 0:
            break
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    print(f"Root = {c:.6f}")

bisection_method()
```
