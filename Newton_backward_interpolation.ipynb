```python
import numpy as np

def newton_backward_interpolation():
    # Get the number of data points from the user
    n = int(input("Enter number of data points: "))

    # Initialize numpy arrays to store x and y values
    x = np.zeros(n)
    y = np.zeros(n)

    print("Enter x and y values for each data point:")
    for i in range(n):
        # Read x and y values for each point interactively
        try:
            x[i], y[i] = map(float, input(f"Enter x[{i}] y[{i}] (space-separated): ").split())
        except ValueError:
            print("Invalid input. Please enter two numbers separated by a space.")
            return # Exit if input is invalid

    # Initialize the backward difference table as a 2D numpy array
    # d[i, j] will store the j-th backward difference of y_i
    d = np.zeros((n, n))
    # The first column of the difference table is the y-values
    d[:, 0] = y

    # Calculate the backward difference table
    # This loop populates the difference table column by column
    for j in range(1, n): # Iterate through columns (order of difference)
        for i in range(n - 1, j - 1, -1): # Iterate from the last row up to j
            d[i, j] = d[i, j-1] - d[i-1, j-1]

    # Get the value of x for which to interpolate
    xp = float(input("Enter the value of x for interpolation (xp): "))

    # Calculate h, the step size (assuming equidistant points)
    h = x[1] - x[0]

    # Calculate 'p' for Newton's Backward Interpolation formula
    # p = (xp - x_n) / h, where x_n is the last x-value (x[n-1])
    p = (xp - x[n-1]) / h

    # Initialize the sum with the last y-value (y[n-1])
    interpolated_value = y[n-1]

    # Calculate the interpolation using the backward difference formula
    term_product = 1.0 # This term accumulates p * (p+1) * ... / i!
    for i in range(1, n):
        # Update the product term for the current iteration
        term_product = term_product * (p + (i - 1)) / i
        # Add the current term to the interpolated value
        # d[n-1, i] is the i-th backward difference of y at the last point
        interpolated_value += term_product * d[n-1, i]

    print(f"\nInterpolated value at x = {xp:.4f} is: {interpolated_value:.6f}")

# Run the interpolation function
newton_backward_interpolation()
```
