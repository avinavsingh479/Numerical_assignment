```python
def f(x):
    return x*x*x - x - 2

def regula_falsi_method():
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    n = int(input("Enter iterations: "))

    c = 0.0 # Initialize c

    for i in range(n):
        fa = f(a)
        fb = f(b)

        # Check for division by zero
        if (fb - fa) == 0:
            print("Error: Denominator (f(b) - f(a)) is zero. Cannot proceed with Regula Falsi.")
            return

        c = (a * fb - b * fa) / (fb - fa)
        fc = f(c)

        if fc == 0:
            break
        if fa * fc < 0:
            b = c
        else:
            a = c

    print(f"Root = {c:.6f}")

regula_falsi_method()
```
